##Bubble Sort

bubble sort.png



#### **Overview:**
Bubble Sort is one of the simplest sorting algorithms. It repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. This process is repeated until the entire list is sorted. It is called "Bubble Sort" because larger elements "bubble" up to the top (or end) of the list during each pass.

#### **How Bubble Sort Works:**

1. **Initial Setup:**
   Start at the first element of the array.
   
2. **Comparison of Adjacent Elements:**
   Compare the current element with the next element in the array.
   
3. **Swapping Elements:**
   If the current element is greater than the next element, swap the two elements. This ensures that the larger element moves towards the right side of the array (or the "end").

4. **Repeat for All Elements:**
   After one pass through the array, the largest element will have bubbled up to its correct position at the end of the array.

5. **Reduce the Range:**
   After the first pass, the last element is sorted. The algorithm then reduces the range of elements it needs to process in the next iteration (since the largest element is in its correct position).

6. **Repeat the Process:**
   The algorithm repeats the process of comparing and swapping adjacent elements for the remaining unsorted part of the array. This continues until no swaps are needed, indicating that the array is fully sorted.

#### **Detailed Step-by-Step Process:**

Consider the array `[5, 3, 8, 4, 2]`:

- **First Pass:**
  - Compare `5` and `3`: Since `5 > 3`, swap them → `[3, 5, 8, 4, 2]`
  - Compare `5` and `8`: No swap needed → `[3, 5, 8, 4, 2]`
  - Compare `8` and `4`: Since `8 > 4`, swap them → `[3, 5, 4, 8, 2]`
  - Compare `8` and `2`: Since `8 > 2`, swap them → `[3, 5, 4, 2, 8]`

  After the first pass, the largest number `8` is at the end.

- **Second Pass:**
  - Compare `3` and `5`: No swap needed → `[3, 5, 4, 2, 8]`
  - Compare `5` and `4`: Since `5 > 4`, swap them → `[3, 4, 5, 2, 8]`
  - Compare `5` and `2`: Since `5 > 2`, swap them → `[3, 4, 2, 5, 8]`

  After the second pass, the next largest number `5` is in its correct position.

- **Third Pass:**
  - Compare `3` and `4`: No swap needed → `[3, 4, 2, 5, 8]`
  - Compare `4` and `2`: Since `4 > 2`, swap them → `[3, 2, 4, 5, 8]`

  After the third pass, the array is still not sorted completely, but `4` is now in its correct position.

- **Fourth Pass:**
  - Compare `3` and `2`: Since `3 > 2`, swap them → `[2, 3, 4, 5, 8]`

  Now, the array is completely sorted.

#### **Key Characteristics:**

1. **Time Complexity:**
   - **Best case**: O(n) when the list is already sorted or nearly sorted (optimized version). In this case, only one pass is needed to confirm the array is sorted.
   - **Worst case**: O(n²) when the array is in reverse order. Each pass requires n-1 comparisons, and this process is repeated n times.
   - **Average case**: O(n²) in general, as the number of comparisons is quadratic.
   
2. **Space Complexity:**
   - **O(1)**: Bubble Sort is an **in-place** sorting algorithm, meaning it doesn't require additional memory apart from the input array. It sorts the array by modifying the original elements directly.

3. **Stability:**
   - **Stable Sort**: Yes. If two elements have the same value, their relative order remains unchanged after sorting. This is important when sorting objects that contain multiple fields (e.g., sorting a list of people by name while keeping the original order based on age).

4. **Adaptiveness:**
   - **Adaptive**: In the best case, Bubble Sort can perform better if the list is already partially sorted, because it can terminate early when no swaps are needed in a pass.

5. **Swapping Mechanism:**
   - Bubble Sort swaps elements whenever they are out of order. This ensures that the largest unsorted element always moves to the end of the list (or top of the list, depending on the direction of sorting) after each pass.
  
#### **Advantages of Bubble Sort:**

- **Simplicity**: The algorithm is very simple and easy to understand, making it ideal for educational purposes.
- **In-place Sorting**: Bubble Sort does not require extra space, which makes it memory efficient.

#### **Disadvantages of Bubble Sort:**

- **Inefficient for Large Datasets**: Its worst and average-case time complexity of O(n²) makes it unsuitable for large datasets. As the size of the input grows, the performance of Bubble Sort deteriorates significantly.
- **Slow Performance**: Bubble Sort is slower compared to more efficient algorithms such as Quick Sort, Merge Sort, or Heap Sort, especially when dealing with large or unsorted data.

#### **When to Use Bubble Sort:**

- **Small Datasets**: Bubble Sort might be used for small datasets, where its inefficiencies are less noticeable.
- **Educational Use**: It's a great algorithm for teaching the basics of sorting because it’s simple and easy to understand.
- **Nearly Sorted Data**: If the data is already sorted or nearly sorted, Bubble Sort can perform quite efficiently.

#### **Optimized Bubble Sort:**
The classic Bubble Sort can be optimized by introducing a flag that tracks whether any swaps were made during a pass. If no swaps are made, the algorithm can terminate early, improving the best-case time complexity to O(n). This optimization ensures that the algorithm performs better on partially sorted or already sorted arrays.

````python
class BubbleSort:
  def __init__(self):
    self.arr = []

  def insert(self):
    number = int(input("Enter the number of the values in an array:"))
    for i in range(0,number):
      n = int(input(f"Enter the element at {i}:"))
      self.arr.append(n)
    print()
    print("Your Array:")
    print(self.arr)

  def sort(self):
    n = len(self.arr)
    for i in range(n):
      flag = False
      for j in range(0 , n-i-1):
        if self.arr[j] > self.arr[j+1]:
          self.arr[j] , self.arr[j+1] = self.arr[j+1] , self.arr[j]
          flag = True

      if not flag:
        break
    print()
    print("Your Array after sorting:")
    print(self.arr)


  def main(self):
    while True:
      print("""
            1. Insert the element
            2. Sort the array
            3. Exit""")

      choice = int(input("Choose Your preference: "))
      if choice == 1:
        self.insert()

      elif choice == 2:
        self.sort()

      elif choice == 3:
        break

      else:
        print("INVALID INPUT")

Sort = BubbleSort()
Sort.main()
````

##Selection Sort


selection.webp


#### **Overview:**
Selection Sort is another simple comparison-based sorting algorithm. It works by repeatedly selecting the minimum (or maximum, depending on the sorting order) element from the unsorted portion of the array and swapping it with the first unsorted element. The process continues until the entire array is sorted.

#### **How Selection Sort Works:**

1. **Initial Setup:**
   - Start with the first element of the array as the **minimum**.

2. **Find the Minimum:**
   - Iterate through the unsorted portion of the array to find the minimum element.

3. **Swap the Minimum with the First Unsorted Element:**
   - Once the minimum element is found, swap it with the first element of the unsorted portion of the array.

4. **Repeat:**
   - After each pass, the first unsorted element is now sorted, so reduce the range of the unsorted part of the array. Repeat the process until the entire array is sorted.

#### **Detailed Step-by-Step Process:**

Consider the array `[64, 25, 12, 22, 11]`.

- **First Pass:**
  - The unsorted portion is `[64, 25, 12, 22, 11]`.
  - Find the minimum: `11`.
  - Swap `11` with the first element (`64`).
  - The array after the first pass: `[11, 25, 12, 22, 64]`.

- **Second Pass:**
  - The unsorted portion is `[25, 12, 22, 64]`.
  - Find the minimum: `12`.
  - Swap `12` with the first element of the unsorted part (`25`).
  - The array after the second pass: `[11, 12, 25, 22, 64]`.

- **Third Pass:**
  - The unsorted portion is `[25, 22, 64]`.
  - Find the minimum: `22`.
  - Swap `22` with `25`.
  - The array after the third pass: `[11, 12, 22, 25, 64]`.

- **Fourth Pass:**
  - The unsorted portion is `[25, 64]`.
  - Find the minimum: `25` (it's already in place).
  - The array after the fourth pass: `[11, 12, 22, 25, 64]`.

- **Fifth Pass:**
  - The array is fully sorted: `[11, 12, 22, 25, 64]`.

#### **Key Characteristics:**

1. **Time Complexity:**
   - **Best Case:** O(n²) – In Selection Sort, even if the array is already sorted, it still needs to go through the whole array and find the minimum element in each pass.
   - **Worst Case:** O(n²) – In the worst case, the algorithm will perform the same number of comparisons and swaps as the best case.
   - **Average Case:** O(n²) – In most cases, Selection Sort performs a quadratic number of comparisons and swaps.
   
2. **Space Complexity:**
   - **O(1)** – Selection Sort is an **in-place** sorting algorithm, meaning it does not require any additional space beyond the input array.

3. **Stability:**
   - **Not Stable** – Selection Sort is not a stable sort because equal elements might not retain their relative order after the sorting process.

4. **Swapping:**
   - **Number of Swaps:** Selection Sort minimizes the number of swaps compared to other sorting algorithms (such as Bubble Sort), as it only swaps when it finds the minimum element.

5. **Adaptiveness:**
   - **Not Adaptive:** Selection Sort is not adaptive. It doesn’t take advantage of partially sorted arrays, so it always performs the same number of comparisons regardless of the input.

#### **Advantages of Selection Sort:**

- **Simplicity:** Selection Sort is easy to understand and implement. It’s often used for educational purposes to demonstrate sorting algorithms.
- **In-place Sorting:** It does not require additional memory beyond the input array, making it space-efficient.

#### **Disadvantages of Selection Sort:**

- **Inefficient for Large Datasets:** The time complexity of O(n²) makes it inefficient for large arrays compared to more advanced algorithms like Merge Sort or Quick Sort.
- **Not Stable:** Equal elements might not retain their relative order, which could be a problem when sorting records that need to preserve order based on secondary criteria.

#### **Correct Selection Sort Structure:**

Here’s how the Selection Sort works in a typical implementation:

1. **Outer Loop:**
   - The outer loop keeps track of the boundary of the unsorted portion of the array. Initially, the unsorted portion is the entire array.

2. **Inner Loop:**
   - The inner loop scans through the unsorted portion of the array, finds the smallest (or largest) element, and stores its index.

3. **Swap:**
   - Once the minimum element in the unsorted portion is found, it is swapped with the first unsorted element. This element is now considered sorted.

4. **Repeat:**
   - Repeat the process for the next unsorted portion of the array until the entire array is sorted.

### **Example Walkthrough:**

Consider sorting the array `[64, 25, 12, 22, 11]` in ascending order.

- **First Pass:**
  - Start with the whole array: `[64, 25, 12, 22, 11]`.
  - Find the minimum element, `11`.
  - Swap `11` with `64`.
  - After the first pass: `[11, 25, 12, 22, 64]`.

- **Second Pass:**
  - Now, consider the subarray `[25, 12, 22, 64]`.
  - Find the minimum element, `12`.
  - Swap `12` with `25`.
  - After the second pass: `[11, 12, 25, 22, 64]`.

- **Third Pass:**
  - Consider the subarray `[25, 22, 64]`.
  - Find the minimum element, `22`.
  - Swap `22` with `25`.
  - After the third pass: `[11, 12, 22, 25, 64]`.

- **Fourth Pass:**
  - Consider the subarray `[25, 64]`.
  - `25` is already the smallest, so no swap needed.
  - After the fourth pass: `[11, 12, 22, 25, 64]`.

- **Fifth Pass:**
  - The array is now sorted: `[11, 12, 22, 25, 64]`.



````python
class SelectionSort:
  def __init__(self):
    self.arr = []

  def insert(self):
      number = int(input("Enter the number of the values in an array:"))
      for i in range(0,number):
        n = int(input(f"Enter the element at {i}:"))
        self.arr.append(n)
      print()
      print("Your Array:")
      print(self.arr)

  def sort(self):
      n = len(self.arr)
      for i in range(n):
        min = i
        for j in range(i+1,n):
          if self.arr[j] < self.arr[min]:
            min = j

          self.arr[i] , self.arr[min] = self.arr[min] , self.arr[i]

      print()
      print("Your Array After Sorting:")
      print(self.arr)

  def main(self):
      while True:
        print("""
              1. Insert the element
              2. Sort the array
              3. Exit""")

        choice = int(input("Choose Your preference: "))
        if choice == 1:
          self.insert()

        elif choice == 2:
          self.sort()

        elif choice == 3:
          break

        else:
          print("INVALID INPUT")

sort = SelectionSort()
sort.main()
````

##Quick Sort

qucik.webp



Quick Sort is a **divide-and-conquer** sorting algorithm that is efficient in practice for large datasets. It was developed by Tony Hoare in 1960. It sorts an array by dividing it into smaller sub-arrays, sorting them independently, and then combining the results.

### How Quick Sort Works:

The general idea of Quick Sort is:
1. **Choose a Pivot**: Pick an element from the array as a "pivot." The pivot is the element around which the array will be partitioned.
2. **Partitioning**: Re-arrange the array so that all elements less than the pivot are on the left side, and all elements greater than the pivot are on the right side.
3. **Recursion**: Recursively apply the same logic to the left and right parts (sub-arrays) of the pivot.

### Steps in Detail:
1. **Choose a Pivot**: There are multiple strategies for choosing a pivot, such as:
   - First element
   - Last element
   - Random element
   - Median of the array
2. **Partitioning**:
   - After selecting the pivot, the array is rearranged so that:
     - Elements less than the pivot are on the left.
     - Elements greater than the pivot are on the right.
   - The pivot ends up in its correct sorted position.
3. **Recursive Sorting**:
   - Recursively sort the left and right sub-arrays that are created by partitioning.
4. **Base Case**:
   - When the size of the sub-array is 1 or 0, it's already sorted.

### Key Points:
- **Average Case Time Complexity**: O(n log n)
- **Worst Case Time Complexity**: O(n²) (can happen if the pivot is not chosen wisely)
- **Best Case Time Complexity**: O(n log n)
- **Space Complexity**: O(log n) (due to recursion stack)
- **In-place Sort**: Yes (does not require additional memory for sorting)
- **Stable**: No (but can be modified to be stable)

---

### Quick Sort Algorithm Explanation:

1. **Choose a Pivot**:
   - Commonly, the **last element** or **random element** is chosen as the pivot.

2. **Partitioning the Array**:
   - We partition the array by moving elements smaller than the pivot to the left, and elements greater than the pivot to the right. The pivot is then placed at its correct position in the sorted array.

3. **Recursion**:
   - Apply Quick Sort recursively to the left and right sub-arrays (the elements before and after the pivot).

---

   - Move all elements less than `5` to the left, and elements greater than `5` to the right. After partitioning, the array will look like:
   ```python
   [1, 5, 8, 9, 7, 10]
   ```
   - `5` is now in its correct position at index 1.

3. **Recursively Apply Quick Sort**:
   - Left sub-array: `[1]` (already sorted).
   - Right sub-array: `[8, 9, 7, 10]`.

Repeat this process for each sub-array until the entire array is sorted.

### Code Implementation of Quick Sort:



### Code Explanation:

1. **Base Case**: If the array has one or no elements (`len(arr) <= 1`), it's already sorted, so return it directly.
   
2. **Pivot Selection**: The pivot is selected as the last element (`arr[-1]`).

3. **Partitioning**:
   - The array is partitioned into two sub-arrays:
     - `left`: Contains elements less than or equal to the pivot.
     - `right`: Contains elements greater than the pivot.
     
4. **Recursive Sorting**:
   - The `quick_sort` function is called recursively on the `left` and `right` sub-arrays.
   
5. **Combine**: The final sorted array is the concatenation of:
   - The sorted `left` sub-array,
   - The pivot,
   - The sorted `right` sub-array.

---

### Time Complexity:

- **Best & Average Case**: O(n log n) — Occurs when the pivot splits the array roughly in half each time.
- **Worst Case**: O(n²) — Occurs when the pivot is the smallest or largest element, leading to unbalanced partitions.
- **Optimal Pivot Selection**: The worst-case time complexity can be avoided with good pivot selection strategies like using the **median of three** or **random pivot**.

---

### Example Walkthrough:

Let’s consider the array `arr = [10, 7, 8, 9, 1, 5]`.

1. **First Call**:
   - `pivot = 5`, array partitioned into `left = [1]` and `right = [10, 7, 8, 9]`.
   - `quick_sort([1])` returns `[1]` (base case).
   - `quick_sort([10, 7, 8, 9])` is processed next.

2. **Second Call (on the Right Sub-array)**:
   - `pivot = 9`, array partitioned into `left = [7, 8]` and `right = [10]`.
   - `quick_sort([7, 8])` continues, and returns `[7, 8]`.
   - `quick_sort([10])` returns `[10]` (base case).

3. **Final Combined Result**:
   - Combine results: `[1] + [5] + [7, 8] + [9] + [10] = [1, 5, 7, 8, 9, 10]`.

````python
class QuickSort:
  def __init__(self):
    self.arr = []

  def insert(self):
      number = int(input("Enter the number of the values in an array:"))
      for i in range(0,number):
        n = int(input(f"Enter the element at {i}:"))
        self.arr.append(n)
      print()
      print("Your Array:")
      print(self.arr)

  def quicksort(self,low,high):
      if low < high:

        pi = self.partition(low,high)

        self.quicksort(low,pi-1)
        self.quicksort(pi+1,high)

  def partition(self,low,high):
      pivot = self.arr[high]

      i = low - 1

      for j in range(low,high):
        if self.arr[j] <= pivot:
          i += 1
          self.arr[i] , self.arr[j] = self.arr[j] , self.arr[i]

      self.arr[i+1] , self.arr[high] = self.arr[high] , self.arr[i+1]

      return i + 1


  def SortedArray(self):
    print("Your Sorted Array:")
    print()
    print(self.arr)

  def main(self):
      while True:
        print("""
              1. Insert the element
              2. Sort the array
              3. See the Sorted Array:
              4. Exit""")

        choice = int(input("Choose Your preference: "))
        if choice == 1:
          self.insert()

        elif choice == 2:
          self.quicksort(0,len(self.arr)-1)

        elif choice == 3:
          self.SortedArray()

        elif choice == 4:
          break

        else:
          print("INVALID INPUT")

sort = QuickSort()
sort.main()
````

##Insertion Sort

### Insertion Sort: Explanation, How It Works, and Code

#### What is Insertion Sort?

**Insertion Sort** is a simple comparison-based sorting algorithm. It builds the final sorted array one element at a time by **inserting each element** into its correct position within the sorted portion of the array.

It's often compared to the way people sort playing cards in their hands: each new card is placed in the correct order relative to the ones already in hand.

### **How Insertion Sort Works**

1. **Start with the second element**: Since the first element is already considered sorted by itself, we start with the second element.
2. **Compare the current element with the previous element(s)**:
   - If the current element is smaller than the previous element, we shift the previous elements one position to the right.
   - Keep shifting until we find the correct position for the current element.
3. **Insert the current element in its correct position**: After shifting the necessary elements, the current element is inserted into the appropriate position in the sorted part of the array.
4. **Repeat for all remaining elements**: We repeat this process for every element until the entire array is sorted.



### **Step-by-Step Explanation with Example:**

Let’s use an example array to explain how Insertion Sort works.

#### Example:
```python
arr = [12, 11, 13, 5, 6]
```

#### 1. **Initial Array**:
Before we begin sorting, the array is:
```python
[12, 11, 13, 5, 6]
```

#### 2. **First Iteration (i = 1)**:
- The **key** is `arr[1] = 11`.
- Compare `11` with the element before it (`arr[0] = 12`):
  - Since `11` is smaller than `12`, we **shift** `12` one position to the right.
  - Now the array looks like: `[12, 12, 13, 5, 6]`.
- Insert `11` at the correct position (`arr[0]`).
  - Now the array is: `[11, 12, 13, 5, 6]`.

#### 3. **Second Iteration (i = 2)**:
- The **key** is `arr[2] = 13`.
- Compare `13` with `arr[1] = 12`. Since `13` is greater than `12`, no shifting is needed.
- Insert `13` at the correct position (no change).
  - Array remains: `[11, 12, 13, 5, 6]`.

#### 4. **Third Iteration (i = 3)**:
- The **key** is `arr[3] = 5`.
- Compare `5` with `arr[2] = 13`. Since `5` is smaller, we shift `13` one position to the right:
  - Array becomes: `[11, 12, 13, 13, 6]`.
- Compare `5` with `arr[1] = 12`. Since `5` is smaller, we shift `12` one position to the right:
  - Array becomes: `[11, 12, 12, 13, 6]`.
- Compare `5` with `arr[0] = 11`. Since `5` is smaller, we shift `11` one position to the right:
  - Array becomes: `[11, 11, 12, 13, 6]`.
- Insert `5` at the correct position (`arr[0]`).
  - Now the array is: `[5, 11, 12, 13, 6]`.

#### 5. **Fourth Iteration (i = 4)**:
- The **key** is `arr[4] = 6`.
- Compare `6` with `arr[3] = 13`. Since `6` is smaller, we shift `13` one position to the right:
  - Array becomes: `[5, 11, 12, 13, 13]`.
- Compare `6` with `arr[2] = 12`. Since `6` is smaller, we shift `12` one position to the right:
  - Array becomes: `[5, 11, 12, 12, 13]`.
- Compare `6` with `arr[1] = 11`. Since `6` is smaller, we shift `11` one position to the right:
  - Array becomes: `[5, 11, 11, 12, 13]`.
- Compare `6` with `arr[0] = 5`. Since `6` is greater than `5`, no shifting is needed.
- Insert `6` at the correct position (`arr[1]`).
  - Now the array is: `[5, 6, 11, 12, 13]`.

#### **Final Sorted Array**:
After all iterations, the array is fully sorted:
```python
[5, 6, 11, 12, 13]
```

### **Key Concepts to Understand Insertion Sort**:
1. **Key Element**: In each iteration, the key element (`arr[i]`) is the element being inserted into its correct position in the sorted portion of the array.
2. **Shifting Elements**: Elements greater than the key element are shifted one position to the right to make space for the key element.
3. **Inserting the Key**: Once the correct position for the key element is found, it is inserted into that position, and the array becomes more sorted.
4. **Iterative Process**: The algorithm works by repeatedly inserting elements into a growing sorted sub-array until the entire array is sorted.

### **Why Insertion Sort is Useful**:
- **Efficient for Small Arrays**: Insertion Sort is **efficient for small arrays** or nearly sorted arrays because its time complexity is O(n) in the best case when the array is already sorted.
- **Adaptive**: It adapts well to the data; if the array is nearly sorted, it performs very efficiently.
- **Stable**: Insertion Sort is a **stable sort**, meaning that equal elements retain their relative order (e.g., if two elements are equal, their positions won’t change).

### **Time Complexity of Insertion Sort**:
- **Best Case**: O(n) (when the array is already sorted).
- **Average Case**: O(n²) (when the elements are in random order).
- **Worst Case**: O(n²) (when the array is sorted in reverse order).

### **Space Complexity**:
- **Space Complexity**: O(1), since Insertion Sort is an **in-place** sorting algorithm that doesn't require any additional memory.

### **When to Use Insertion Sort**:
- When the array is **nearly sorted** or small, Insertion Sort can be much faster than more complex algorithms like Quick Sort or Merge Sort.
- For **real-time applications**, where stability is important, and the data is small, Insertion Sort can be a good choice.

### **Summary**:
- **Insertion Sort** builds the final sorted array one element at a time by inserting each element into its correct position in the sorted portion of the array.
- It works by shifting larger elements to the right to make space for the current element.
- Its time complexity is O(n²) in the worst case but can perform in O(n) time for nearly sorted data.
- It's an in-place and stable sorting algorithm, making it a good choice for small datasets and applications requiring stability.



````python
class InsertionSort:
  def __init__(self):
    self.arr = []

  def insert(self):
      number = int(input("Enter the number of the values in an array:"))
      for i in range(0,number):
        n = int(input(f"Enter the element at {i}:"))
        self.arr.append(n)
      print()
      print("Your Array:")
      print(self.arr)

  def sort(self):
    n = len(self.arr)
    for i in range(1,n):
      key = self.arr[i]

      j = i - 1
      while j >= 0 and self.arr[j] > key:
        self.arr[j+1] = self.arr[j]
        j -= 1

      self.arr[j+1] = key

  def sortedarray(self):
    print()
    print("Sorted Array:")
    print(self.arr)

  def main(self):
      while True:
        print("""
              1. Insert the element
              2. Sort the array
              3. See the sorted array
              4. Exit""")

        choice = int(input("Choose Your preference: "))
        if choice == 1:
          self.insert()

        elif choice == 2:
          self.sort()

        elif choice == 3:
          self.sortedarray()

        elif choice == 4:
          break

        else:
          print("INVALID INPUT")

sort = InsertionSort()
sort.main()
````

### **Detailed Explanation of Merge Sort**

Merge Sort is a popular and efficient **divide and conquer** algorithm used to sort an array or list of elements. It works by recursively dividing the array into two halves, sorting each half, and then merging the sorted halves together.

### **Steps of Merge Sort Algorithm:**

1. **Divide**: The array is divided into two halves. If the array contains more than one element, we continue dividing until each sub-array has only one element (a single-element array is considered sorted).
  
2. **Conquer**: Once the array is divided into sub-arrays, each sub-array is sorted individually. If the sub-array has more than one element, merge sort is recursively applied to each sub-array.

3. **Merge**: After sorting the sub-arrays, we merge them back together in a sorted manner to produce the final sorted array. This is done by comparing the elements of each sub-array and combining them in order.
