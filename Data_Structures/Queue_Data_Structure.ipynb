#Queue Data Structure

A **Queue** is a linear data structure that follows the **FIFO (First In, First Out)** principle, meaning the element that is added first is the one that will be removed first. This structure is commonly used in scenarios where order and sequence are crucial, such as task scheduling, customer service queues, or printer job queues.

### Key Characteristics of a Queue
1. **Front and Rear Pointers**:
   - The **front** pointer indicates the position of the element that is next to be removed.
   - The **rear** pointer indicates where the next element will be inserted.
2. **FIFO Order**: Elements are added to the back of the queue and removed from the front.

### Queue Operations
Queues typically support four main operations:

1. **Enqueue (Insertion)**
   - Adds an element to the rear (end) of the queue.
   - If the queue is full (for fixed-size queues), an overflow error occurs.
   - Example: A customer joins the end of a checkout line.

2. **Dequeue (Deletion)**
   - Removes the element from the front of the queue.
   - If the queue is empty, an underflow error occurs.
   - Example: A customer at the front of the checkout line is served and leaves the queue.

3. **Peek (Front)**
   - Retrieves the element at the front of the queue without removing it.
   - This allows viewing of the next element to be dequeued.
   - Example: Checking the next customer in line to be served.

4. **IsEmpty**
   - Checks if the queue contains any elements.
   - Returns `True` if there are no elements, and `False` otherwise.

### Variants of Queues
- **Circular Queue**: Connects the front and rear ends, making the queue circular.
- **Priority Queue**: Elements are ordered by priority instead of FIFO.
- **Deque (Double-Ended Queue)**: Allows insertion and deletion from both ends.

### Working Example
Consider a customer service scenario:
1. `Enqueue`: Each new customer joins at the end of the line.
2. `Dequeue`: The customer at the front is served and leaves the line.
3. `Peek`: The agent checks the first customer in line.
4. `IsEmpty`: The agent checks if any customers are left in the line.


```python
class Queue:
  def __init__(self):
    self.queue = []

  def enqueue(self):
    input_ = int(input("Enter the number of elements to pass into the queue:"))
    for i in range(input_):
      number = int(input(f"Pass the elements into the queue at {i+1}: "))
      self.queue.append(number)
      print()
    print("The elements are successfully added in the queue")

  def dequeue(self):
    if (len(self.queue)) != 0:
      print(f"The element {self.queue[0]} has been dequeued")
      self.queue.pop(0)
    else:
      print("The queue is empty")

  def peek(self):
    if (len(self.queue)) != 0:
      print(f"The element at the first in the queue is {self.queue[0]}")

    else:
      print("The queue is empty")


  def isempty(self):
    if (len(self.queue)) == 0:
      print("The queue is empty")
    else:
      print("The queue is not empty")


  def display(self):
    print("Your Queue:")
    if len(self.queue) != 0:
      for i in self.queue:
        print(i,end = "  ")

  def menu(self):
    while True:
      print("""
            1. Enqueue
            2. Dequeue
            3. peek
            4. isEmpty
            5. display
            6. Exit""")
      print()
      choice = int(input("Chooose Your Preference:"))


      if choice == 1:
        self.enqueue()

      elif choice == 2:
        self.dequeue()

      elif choice == 3:
        self.peek()

      elif choice == 4:
        self.isempty()

      elif choice == 5:
        self.display()

      elif choice == 6:
        break

      else:
        print("Invalid input")

queue = Queue()
queue.menu()

```
