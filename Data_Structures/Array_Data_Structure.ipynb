##One Dimensional array


#### Definition:
- A one-dimensional array is a linear collection of elements stored in contiguous memory locations. It can be thought of as a list of elements.

#### Characteristics:
- **Indexing**: Elements are accessed using a single index.
- **Homogeneous**: All elements must be of the same data type.

#### Operations:
- **Insertion**: Adding an element at a specific index or at the end.
- **Deletion**: Removing an element from a specific index or from the end.
- **Traversal**: Iterating through the elements to access or modify them.
- **Searching**: Finding an element using linear or binary search.

#### Applications:
- Storing collections of data, such as scores in a game or a list of names.
- Implementing other data structures (e.g., stacks and queues).

##Two Dimensional Array



#### Definition:
- A two-dimensional array is essentially an array of arrays, representing data in a grid or table format (rows and columns).

#### Characteristics:
- **Indexing**: Elements are accessed using two indices (row and column).
- **Homogeneous**: All elements must be of the same data type.

#### Operations:
- **Insertion**: Adding an element at a specific row and column.
- **Deletion**: Removing an element from a specific row and column.
- **Traversal**: Iterating through the array in nested loops to access or modify elements.
- **Searching**: Finding an element using various strategies (linear search, binary search on sorted rows/columns).

#### Applications:
- Representing matrices in mathematics.
- Storing image data (pixels in a 2D grid).
- Implementing graphs and game boards (e.g., chess or tic-tac-toe).

```python
class Two_D_Array:
  def __init__(self,rows,cols):
    self.rows = rows
    self.cols = cols
    self.arr = []
    for i in range(rows):
      row = []
      for j in range(cols):
        row.append(0)
      self.arr.append(row)


  def insert(self):
    for i in range (self.rows):
      for j in range (self.cols):
        self.arr[i][j] = int(input(f"Enter the element at position [{i+1}][{j+1}]: "))
        print()

  def traverse(self):
    print("Your 2-D Array:")
    for i in range(self.rows):
      for j in range(self.cols):
        print(self.arr[i][j],end=" ")
      print()


  def delete(self):
    delete = int(input("Enter the element you want to delete: "))
    flag =  False
    for i in range(self.rows):
      for j in range(self.cols):
        if delete == self.arr[i][j]:
          self.arr[i][j] = 0
          flag = True

    if not flag:
      print("The element is not present in the array")

  def search(self):
    search = int(input('Enter the element you want to search: '))
    flag = False
    for i in range(self.rows):
      for j in range(self.cols):
        if search == self.arr[i][j]:
          print(f"Your element {search} is found at index [{i+1}][{j+1}]")
          flag = True
    if not flag:
      print("The element is not found in the array")



  def menu(self):
    while True:
      print("""
            1. Insert
            2. Delete
            3. Search
            4. Traverse
            5. Exit""")

      choice = int(input("Enter Your Preference:"))
      print()

      if choice == 1:
        self.insert()

      elif choice == 2:
        self.delete()

      elif choice == 3:
        self.search()

      elif choice == 4:
        self.traverse()

      else:
        break

row_number = int(input("Enter the number of rows: "))
col_number = int(input('Enter the number of columns: '))

Array = Two_D_Array(row_number, col_number)
Array.menu()
```
