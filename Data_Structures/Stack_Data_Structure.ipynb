### Stack Data Structure

A **stack** is a linear data structure that follows the **Last In, First Out (LIFO)** principle, which means the last element added to the stack is the first one to be removed. You can think of a stack like a stack of plates: you can only add or remove the top plate.

#### Key Characteristics:
- **LIFO Order**: The last element added is the first to be removed.
- **Dynamic Size**: The size can grow or shrink as needed.
- **Access**: Only the top element can be accessed directly.

### Common Operations
1. **Push**: Add an element to the top of the stack.
2. **Pop**: Remove and return the top element from the stack.
3. **Peek/Top**: Return the top element without removing it from the stack.
4. **isEmpty**: Check if the stack is empty.



### Brief Explanation of Operations

1. **Push**:
   - **Operation**: Adds an element to the top.
   - **Example**: Pushing `2` onto the stack `[1]` results in `[1, 2]`.

2. **Pop**:
   - **Operation**: Removes the top element and returns it.
   - **Example**: Popping from `[1, 2]` returns `2`, leaving `[1]`.

3. **Peek**:
   - **Operation**: Returns the top element without removing it.
   - **Example**: Peeking at `[1, 2]` returns `2`, stack remains `[1, 2]`.

4. **isEmpty**:
   - **Operation**: Checks if the stack is empty.
   - **Example**: For an empty stack `[]`, it returns `True`.

### Real-World Applications
- **Function Call Management**: Stacks are used to manage function calls in programming languages, where each function call is pushed onto a stack and popped when completed.
- **Undo Mechanisms**: Applications like text editors often use stacks to implement undo features, where each action can be pushed onto a stack and undone by popping.
- **Expression Evaluation**: Stacks are used in parsing expressions and evaluating mathematical expressions in compilers and interpreters.

```python
class Stack:
  def __init__(self):
    self.stack = []

  def push(self):
    length = int(input("Enter the number of elements you want to insert in the stack :"))
    for i in range(length):
      num = int(input(f"Insert the element in the stack at position {i+1}: "))
      print()
      self.stack.append(num)
    print("Your stack is successfully created")

  def isempty(self):
    if len(self.stack) == 0:
      print("The Stack is Empty")
    else:
      print("The Stack is not Empty")

  def display(self):
    print("Your Stack:")
    for i in reversed(self.stack):
      print(i)

  def pop(self):
    if len(self.stack) != 0:
      print(f"The element {self.stack[0]} is successfully popped")
      self.stack.pop(0)
    else:
      print("Your Stack is empty")

  def peek(self):
    if len(self.stack) != 0:
      print(f"The top element of the stack is {self.stack[0]}")
    else:
      print("Your Stack is empty")
  def menu(self):
    while True:
      print("""
            1. Push
            2. Pop
            3. IsEmpty
            4. Peek
            5. Display
            6. Exit""")
      print()
      choice = int(input("Enter your choice:"))
      print()
      if choice == 1:
        self.push()

      elif choice == 2:
        self.pop()

      elif choice == 3:
        self.isempty()

      elif choice == 4:
        self.peek()

      elif choice == 5:
        self.display()

      else:
        break

stack = Stack()
stack.menu()
```
