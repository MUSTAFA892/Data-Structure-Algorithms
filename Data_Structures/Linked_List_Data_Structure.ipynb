# Linked List Data Structure

There are several types of linked lists, each with its own structure and use cases. The main types include:

1. **Singly Linked List**:
   - Each node contains data and a pointer to the next node.
   - Traversal is only in one direction (forward).

2. **Doubly Linked List**:
   - Each node contains data, a pointer to the next node, and a pointer to the previous node.
   - Allows traversal in both directions (forward and backward).

3. **Circular Linked List**:
   - The last node points back to the first node, forming a circle.
   - Can be singly or doubly circular:
     - **Singly Circular Linked List**: Nodes point to the next node, with the last node pointing to the first.
     - **Doubly Circular Linked List**: Nodes have pointers to both the next and previous nodes, with the last node linking back to the first.

4. **Multilevel Linked List**:
   - Each node can point to another linked list, allowing for a hierarchical structure. This type is useful for representing complex data structures like tree structures.

5. **Skip List**:
   - A linked list with multiple layers, where some nodes are "skipped" to allow faster search times. Each node in the higher layers points to nodes in the lower layers.


# Singly linked list

A **singly linked list** is a type of linked list where each node contains two components: the data and a pointer to the next node in the sequence. This structure allows for efficient insertion and deletion of nodes, as the elements are not stored in contiguous memory locations.

#### Structure of a Singly Linked List

Each node in a singly linked list has the following structure:

```plaintext
Node {
    Data      // The value or information stored in the node
    Next      // A pointer/reference to the next node in the list
}
```

The list is typically defined by a **head** pointer, which points to the first node. If the list is empty, the head pointer is `null`.

#### Operations on a Singly Linked List

1. **Insertion**:
   - **At the Beginning**:
     - Create a new node.
     - Set the new node's `Next` pointer to the current head.
     - Update the head to point to the new node.

   - **At the End**:
     - Create a new node.
     - Traverse the list to find the last node (where `Next` is `null`).
     - Set the last node's `Next` pointer to the new node.

   - **At a Specific Position**:
     - Create a new node.
     - Traverse the list to find the node just before the desired position.
     - Adjust pointers to insert the new node.

2. **Deletion**:
   - **From the Beginning**:
     - Update the head to point to the second node (the node after the head).

   - **From the End**:
     - Traverse the list to find the second-to-last node.
     - Set its `Next` pointer to `null`, effectively removing the last node.

   - **From a Specific Position**:
     - Traverse to the node just before the desired position.
     - Adjust pointers to skip the node to be deleted.

3. **Traversal**:
   - Start from the head and continue to follow the `Next` pointers until you reach a node where `Next` is `null`.
   - This allows you to access each node in the list.

4. **Searching**:
   - Start from the head and traverse the list, comparing each node's data with the target value until the node is found or the end of the list is reached.

5. **Reversing**:
   - Change the direction of the `Next` pointers so that they point to the previous node instead of the next one.

#### Advantages of Singly Linked Lists

- **Dynamic Size**: Unlike arrays, singly linked lists can grow or shrink in size dynamically.
- **Efficient Insertions/Deletions**: Adding or removing nodes can be done in constant time, especially at the head.

#### Disadvantages of Singly Linked Lists

- **No Backward Traversal**: You can only traverse the list in one direction.
- **Memory Overhead**: Each node requires additional memory for the pointer.
- **Sequential Access**: You must traverse the list to access elements, which can be slower than direct access in arrays.

### Doubly Linked List

A **doubly linked list** is a type of linked list where each node contains three components: the data, a pointer to the next node, and a pointer to the previous node. This structure allows for more flexible traversal of the list, as you can move in both directions (forward and backward).

#### Structure of a Doubly Linked List

Each node in a doubly linked list has the following structure:

```plaintext
Node {
    Prev      // A pointer/reference to the previous node in the list
    Data      // The value or information stored in the node
    Next      // A pointer/reference to the next node in the list
}
```

The list is typically defined by a **head** pointer, which points to the first node, and a **tail** pointer, which points to the last node. If the list is empty, both head and tail pointers are `null`.

#### Operations on a Doubly Linked List

1. **Insertion**:
   - **At the Beginning**:
     - Create a new node.
     - Set the new node's `Next` pointer to the current head.
     - Set the current head's `Prev` pointer to the new node (if head is not null).
     - Update the head to point to the new node.

   - **At the End**:
     - Create a new node.
     - Set the new node's `Prev` pointer to the current tail.
     - Set the current tail's `Next` pointer to the new node (if tail is not null).
     - Update the tail to point to the new node.

   - **At a Specific Position**:
     - Create a new node.
     - Traverse the list to find the node just before the desired position.
     - Adjust the pointers of the previous and next nodes to insert the new node.

2. **Deletion**:
   - **From the Beginning**:
     - Update the head to point to the second node (the node after the head).
     - Set the new head's `Prev` pointer to `null`.

   - **From the End**:
     - Traverse the list to find the second-to-last node.
     - Update its `Next` pointer to `null`, effectively removing the last node.
     - Set the new tail's `Next` pointer to `null`.

   - **From a Specific Position**:
     - Traverse to the node at the desired position.
     - Adjust the pointers of the previous and next nodes to skip the node to be deleted.

3. **Traversal**:
   - Start from the head and follow the `Next` pointers until you reach a node where `Next` is `null` (forward traversal).
   - Start from the tail and follow the `Prev` pointers until you reach a node where `Prev` is `null` (backward traversal).

4. **Searching**:
   - Start from the head and traverse the list, comparing each node's data with the target value until the node is found or the end of the list is reached.
   - Searching can also be done backward from the tail.

5. **Reversing**:
   - Change the direction of both `Next` and `Prev` pointers for each node, effectively reversing the list.

#### Advantages of Doubly Linked Lists

- **Bidirectional Traversal**: You can traverse the list in both forward and backward directions.
- **Easier Deletion**: In deletion operations, you have a direct pointer to the previous node, making it simpler to adjust pointers.
- **Flexibility**: Can easily insert or delete nodes from both ends of the list.

#### Disadvantages of Doubly Linked Lists

- **Memory Overhead**: Each node requires additional memory for two pointers (`Prev` and `Next`), increasing the overall memory usage.
- **Complexity**: The implementation is generally more complex than a singly linked list due to the need to manage two pointers per node.
- **Slightly Slower**: Due to the extra pointers, operations may be slightly slower compared to singly linked lists, especially in terms of memory access.

#Circular Linked List

A **circular linked list** is a variation of a linked list in which the last node points back to the first node, forming a circle. This structure eliminates the null references at the end of the list, allowing for continuous traversal from any node.

### Structure of a Circular Linked List

Each node in a circular linked list has the following structure:

```plaintext
Node {
    Data      // The value or information stored in the node
    Next      // A pointer/reference to the next node in the list
}
```

In this structure, a single pointer (usually referred to as the head) is maintained to keep track of the starting point of the list.

### Operations on a Circular Linked List

1. **Insertion**:
   - **At the Beginning**:
     - Create a new node.
     - If the list is empty, set the new node's `Next` pointer to point to itself.
     - If the list is not empty, set the new node's `Next` pointer to the current head and traverse to the last node to update its `Next` pointer to the new node.
     - Update the head to point to the new node.

   - **At the End**:
     - Create a new node.
     - If the list is empty, set the new node's `Next` pointer to point to itself.
     - If the list is not empty, traverse to the last node, set its `Next` pointer to the new node, and set the new node's `Next` pointer to the head.

   - **At a Specific Position**:
     - Create a new node.
     - Traverse the list to find the node just before the desired position.
     - Adjust the pointers to insert the new node.

2. **Deletion**:
   - **From the Beginning**:
     - If the list has only one node, set the head to `null`.
     - If there are multiple nodes, update the head to point to the second node and traverse to the last node to update its `Next` pointer to the new head.

   - **From the End**:
     - Traverse to find the second-to-last node and set its `Next` pointer to the head, effectively removing the last node.

   - **From a Specific Position**:
     - Traverse to the node at the desired position.
     - Adjust the pointers to skip the node to be deleted.

3. **Traversal**:
   - Start from the head and follow the `Next` pointers until you return to the head.

4. **Searching**:
   - Start from the head and traverse the list using `Next` pointers, comparing each node's data with the target value until the node is found or you return to the head.

5. **Reversing**:
   - Reversing a circular linked list is more complex and may require rearranging pointers to maintain the circular structure.

### Advantages of Circular Linked Lists

- **Continuous Traversal**: You can traverse the list starting from any node and will never reach a null pointer, allowing for easier iteration.
- **Efficient Memory Usage**: Does not require sentinel nodes (like a dummy node), making it memory efficient for certain applications.
- **Flexible**: Can be used in applications where the list needs to be traversed repeatedly without stopping.

### Disadvantages of Circular Linked Lists

- **Complexity**: Implementation and handling of edge cases (like insertion and deletion) can be more complex than linear linked lists.
- **Infinite Loops**: Care must be taken during traversal to avoid infinite loops if the traversal logic is not correctly implemented.
- **No Tail Pointer**: Without a tail pointer, operations that require access to the end of the list can be inefficient.



````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = new_node
            new_node.next = self.head
            self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = new_node
            new_node.next = self.head

    def traverse(self):
        if not self.head:
            print("Your Circular Linked List is empty.")
            return
        print("Your Circular Linked List:")
        current = self.head
        while True:
            print(current.data, end=" -> ")
            current = current.next
            if current == self.head:
                break
        print("(Back To Head)")

    def search(self, target):
        if not self.head:
            print("The list is empty.")
            return
        current = self.head
        count = 1
        while True:
            if current.data == target:
                print(f"The target value {target} found at node: {count}")
                return
            current = current.next
            count += 1
            if current == self.head:
                break
        print(f"The target value {target} is not found in the CLL")

    def delete(self, target):
        if not self.head:
            print("The list is empty.")
            return
        current = self.head
        previous = None

        if current.data == target:
            if current.next == self.head:
                self.head = None
            else:
                while current.next != self.head:
                    current = current.next
                current.next = self.head.next
                self.head = self.head.next
            print(f"The target value {target} is successfully deleted")
            return

        while current.next != self.head:
            if current.next.data == target:
                current.next = current.next.next
                print(f"The target value {target} is successfully deleted")
                return
            current = current.next

        print(f"The target value {target} is not found in the CLL")

    def no_of_nodes(self):
        if not self.head:
            print("The list is empty.")
            return 0
        current = self.head
        count = 0
        while True:
            count += 1
            current = current.next
            if current == self.head:
                break
        print(f"The total number of nodes in CLL: {count}")

    def menu(self):
        while True:
            print("""
                1. Insert at the beginning
                2. Insert at the End
                3. Traverse
                4. Search
                5. Delete
                6. Number of nodes
                7. Exit""")
            choice = int(input("Choose Your Preferences: "))

            if choice == 1:
                input_ = int(input("Enter the number of nodes you want to insert at the beginning: "))
                for _ in range(input_):
                    number = int(input("Enter the number: "))
                    self.insert_at_beginning(number)
                print("Your values are inserted at the beginning")

            elif choice == 2:
                input_ = int(input("Enter the number of nodes you want to insert at the end: "))
                for _ in range(input_):
                    number = int(input("Enter the number: "))
                    self.insert_at_end(number)
                print("Your values are inserted at the end")

            elif choice == 3:
                self.traverse()

            elif choice == 4:
                search = int(input("Enter the element you want to search: "))
                self.search(search)

            elif choice == 5:
                delete = int(input("Enter the element you want to delete: "))
                self.delete(delete)

            elif choice == 6:
                self.no_of_nodes()

            elif choice == 7:
                break

            else:
                print("INVALID INPUT :(")

CLL = CircularLinkedList()
CLL.menu()
````

# Doubly Circular Linked List

A **doubly circular linked list** is a variation of the doubly linked list where the last node's `Next` pointer points back to the first node, and the first node's `Prev` pointer points back to the last node. This structure allows for circular traversal in both directions.

### Structure of a Doubly Circular Linked List

Each node in a doubly circular linked list has the following structure:

```plaintext
Node {
    Prev      // A pointer/reference to the previous node in the list
    Data      // The value or information stored in the node
    Next      // A pointer/reference to the next node in the list
}
```

In this structure, both the head and tail nodes are part of the same circular structure, and you can reach any node from any other node by continuously following the `Next` or `Prev` pointers.

### Operations on a Doubly Circular Linked List

1. **Insertion**:
   - **At the Beginning**:
     - Create a new node.
     - If the list is empty, set the new node's `Next` and `Prev` pointers to point to itself.
     - If the list is not empty, set the new node's `Next` to the current head and `Prev` to the current tail.
     - Update the current head's `Prev` pointer to the new node and the current tail's `Next` pointer to the new node.
     - Update the head to point to the new node.

   - **At the End**:
     - Create a new node.
     - If the list is empty, set the new node's `Next` and `Prev` pointers to point to itself.
     - If the list is not empty, set the new node's `Prev` to the current tail and `Next` to the current head.
     - Update the current tail's `Next` pointer to the new node and the current head's `Prev` pointer to the new node.
     - Update the tail to point to the new node.

   - **At a Specific Position**:
     - Create a new node.
     - Traverse the list to find the node just before the desired position.
     - Adjust the pointers of the previous and next nodes to insert the new node.

2. **Deletion**:
   - **From the Beginning**:
     - If the list has only one node, set head and tail to `null`.
     - If there are multiple nodes, update the head to point to the second node, and set the new head's `Prev` pointer to the tail. Update the tail's `Next` pointer to the new head.

   - **From the End**:
     - If the list has only one node, set head and tail to `null`.
     - If there are multiple nodes, find the second-to-last node, update its `Next` pointer to the head, and set the new tail's `Next` pointer to the head.

   - **From a Specific Position**:
     - Traverse to the node at the desired position.
     - Adjust the pointers of the previous and next nodes to skip the node to be deleted.

3. **Traversal**:
   - **Forward Traversal**: Start from the head and follow the `Next` pointers until you return to the head.
   - **Backward Traversal**: Start from the tail and follow the `Prev` pointers until you return to the tail.

4. **Searching**:
   - Start from the head and traverse the list using `Next` pointers, comparing each node's data with the target value until the node is found or you return to the head.

5. **Reversing**:
   - Change the direction of both `Next` and `Prev` pointers for each node, effectively reversing the circular list.

### Advantages of Doubly Circular Linked Lists

- **Circular Traversal**: You can traverse the list in a circular manner without reaching an end.
- **Bidirectional Traversal**: Like the doubly linked list, you can move both forwards and backwards.
- **Efficient Insertions and Deletions**: Insertion and deletion operations can be efficiently performed at both ends without needing special cases for empty lists.

### Disadvantages of Doubly Circular Linked Lists

- **Memory Overhead**: Each node still requires additional memory for two pointers, increasing overall memory usage.
- **Complexity**: The implementation is generally more complex than both singly linked lists and doubly linked lists due to circular connections.
- **Potential for Infinite Loops**: If not carefully handled, traversal operations may lead to infinite loops since the list is circular.

In summary, doubly circular linked lists combine the benefits of both doubly linked lists and circular lists, providing efficient bidirectional traversal with circular structure.

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyCircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            current = self.head.prev
            new_node.next = self.head
            new_node.prev = current
            current.next = new_node
            self.head.prev = new_node
            self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            current = self.head.prev
            current.next = new_node
            new_node.prev = current
            new_node.next = self.head
            self.head.prev = new_node

    def traverse_forward(self):
        if not self.head:
            print("The list is empty.")
            return
        print("Forward Traversal:")
        current = self.head
        while True:
            print(current.data, end=" <-> ")
            current = current.next
            if current == self.head:
                break
        print("(Back to Head)")

    def traverse_backward(self):
        if not self.head:
            print("The list is empty.")
            return
        print("Backward Traversal:")
        current = self.head.prev
        while True:
            print(current.data, end=" <-> ")
            current = current.prev
            if current.next == self.head:
                break
        print("(Back to Head)")

    def delete(self, target):
        if not self.head:
            print("The list is empty.")
            return
        current = self.head
        while True:
            if current.data == target:
                if current == self.head and current.next == self.head:
                    self.head = None
                else:
                    current.prev.next = current.next
                    current.next.prev = current.prev
                    if current == self.head:
                        self.head = current.next
                print(f"The target value {target} is successfully deleted")
                return
            current = current.next
            if current == self.head:
                break
        print(f"The target value {target} is not found in the list")

    def no_of_nodes(self):
        if not self.head:
            print("The list is empty.")
            return 0
        count = 0
        current = self.head
        while True:
            count += 1
            current = current.next
            if current == self.head:
                break
        print(f"The total number of nodes in the list: {count}")
        return count

    def menu(self):
        while True:
            print("""
                1. Insert at the beginning
                2. Insert at the end
                3. Traverse forward
                4. Traverse backward
                5. Delete
                6. Number of nodes
                7. Exit""")
            choice = int(input("Choose Your Preferences: "))
            if choice == 1:
                number = int(input("Enter the number: "))
                self.insert_at_beginning(number)
            elif choice == 2:
                number = int(input("Enter the number: "))
                self.insert_at_end(number)
            elif choice == 3:
                self.traverse_forward()
            elif choice == 4:
                self.traverse_backward()
            elif choice == 5:
                target = int(input("Enter the element you want to delete: "))
                self.delete(target)
            elif choice == 6:
                self.no_of_nodes()
            elif choice == 7:
                break
            else:
                print("INVALID INPUT :(")

DCLL = DoublyCircularLinkedList()
DCLL.menu()
````

## Deep Understanding of Linked List

##  What is a Linked List?

A **Linked List** is a linear data structure where each element (**node**) contains:

* Some **data** (like 5, 10, etc.)
* A **reference (pointer)** to the next node in the list

Think of it like a **chain** of boxes, where each box has:

* A number (data)
* A string tied to the next box (next)

---

##  Node Class (The Box)

```python
class Node:
    def __init__(self, data):
        self.data = data        # Stores the value
        self.next = None        # Points to the next Node (initially None)
```

Every time you do:

```python
node1 = Node(5)
```

It creates:

```
[node1]  data = 5, next = None
```

>  **"How does Python store `self.data` and `self.next` inside each `Node`, and how does it remember all this without an array?"**

---

###  1. Understanding the Node Structure in Memory

When you do:

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

Each `Node` object in Python is like a **dictionary-like object in memory**. It stores:

* The value (`self.data`)
* A reference to the next node (`self.next`)

Example:

```python
node = Node(10)
```

This creates an object in memory that looks like:

```python
node = {
    'data': 10,
    'next': None
}
```

 If we later do:

```python
node2 = Node(20)
node2.next = node
```

Now `node2` looks like:

```python
node2 = {
    'data': 20,
    'next': node
}
```

So `node2.next` is pointing to the memory location of `node`.

---

###  2. No Array? How Does It Remember Everything?

It **doesnt store everything in one block** like an array does.
Instead, it uses **individual objects**, and each object **links to the next one**.

Think of each `Node` like a book page with a turn to page X at the bottom.

* You don't need all pages in one stack.
* Each page tells you where to go next.

This is how linked lists work  every node knows only:

* Its **own value**
* **Where the next node is**

So the memory looks something like:

```
[Node1]  [Node2]  [Node3]  None
```

Each node is stored **anywhere in memory**  they are not side-by-side like arrays.

---

###  Comparison: Array vs Linked List

| Feature             | Array                      | Linked List                     |
| ------------------- | -------------------------- | ------------------------------- |
| Memory layout       | Continuous block           | Scattered nodes                 |
| Stores all elements | Together in one place      | One at a time, each with `next` |
| Access by index     | Fast: `arr[2]`             | Slow: need to follow links      |
| Insert in middle    | Expensive: shift elements  | Easy: change links              |
| Size                | Fixed or resizing required | Dynamic, can grow/shrink easily |

---

###  Internal Working of `self.data` and `self.next`

In CPython (Python's main implementation), every object  including a class  is backed by a **C structure** in memory. When you write:

```python
self.data = 10
```

It:

* Adds a new entry to the objects internal dictionary
* Stores `10` as a reference (not the actual value)
* Python's memory manager keeps track of these

So Python knows:

* The variable `self.data` belongs to this object
* The variable `self.next` refers to another object

---

###  How is traversal possible?

When you write:

```python
current = head
while current:
    print(current.data)
    current = current.next
```

You're telling Python:

1. Start at the `head` (first node)
2. Print its value
3. Jump to `next` (because you stored that link inside the object)
4. Repeat until `next` is `None`

This is exactly how a human would walk through a train: from one carriage to the next.

---

###  Bonus: Behind the Scenes

You can visualize it in memory like this (pseudocode):

```
Node(id=abc1): data=10, next=None
Node(id=abc2): data=20, next=abc1
Node(id=abc3): data=30, next=abc2
```

Your `head` points to `abc3`.

---

##  TL;DR

* `self.data` stores the actual value in the object.
* `self.next` stores a **reference to another Node**, not the value.
* Python uses **objects and references** (like pointers) to create the chain.
* Even though there's no array, Python's memory manager keeps track of where each node is.
* Thats why you can walk from node to node using `.next`.
