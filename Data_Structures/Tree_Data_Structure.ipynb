A **normal tree data structure** is a hierarchical structure consisting of nodes, where each node contains some data and a list of references (or pointers) to its child nodes. Unlike a **binary tree** (where each node has at most two children), a **normal tree** can have an arbitrary number of children for each node.

### Implementation of a Normal Tree

We'll implement a basic tree structure using Python classes, where:
- Each **Node** will hold:
  - Some data (`val`).
  - A list of child nodes (`children`).
- The **Tree** class will just hold the root node and provide some utility methods.

### Steps:
1. **Node class**: Represents each node of the tree.
2. **Tree class**: Represents the tree, including operations to add children, traverse the tree, and more.



### Explanation:

1. **Node Class**:
   - **`val`**: Stores the data associated with the node.
   - **`children`**: A list that holds references to the child nodes of this node.
   - **`add_child` method**: Adds a child node to the `children` list.
   - **`__repr__` method**: Provides a string representation of a node for better display.

2. **Tree Class**:
   - **`root`**: A reference to the root node of the tree.
   - **`add_child` method**: Adds a child node to a parent node (can be the root or any other node).
   - **`display` method**: Displays the tree in a hierarchical structure starting from a given node (default is the root). It uses recursion to print each node and its children with indentation based on the level of depth.

### Operations You Can Implement:
1. **Adding nodes**: You can add nodes to any parent node using `add_child`.
2. **Traversing the tree**: We use a recursive function in `display` to traverse the tree and print it in a hierarchical form. You can also implement other traversal methods such as:
   - Depth-First Search (DFS)
   - Breadth-First Search (BFS)
3. **Searching for a value**: You can implement a method to search for a node with a specific value (using DFS or BFS).
4. **Removing nodes**: You can implement functionality to remove nodes and handle re-linking of the parent-child relationships.

This implementation is quite flexible and can be extended to include more operations or variations, depending on your specific use case.

````python
class Node:
  def __init__(self,val):
    self.val = val
    self.children = []

  def add_child(self,child_node):
    self.children.append(child_node)

  def remove_child(self,child_val):
    self.children = [child for child in self.children if child.val != child_val]

class Tree:
  def __init__(self,root_val):
    self.root = Node(root_val)

  def add_child(self,parent_val,child_val):
    parent_node = self.search(self.root,parent_val)
    if parent_node:
      parent_node.add_child(Node(child_val))
      print(f"Added child {child_val} to parent {parent_val}")
    else:
      print(f"Parent node {parent_val} not found")

  def search(self, node ,target):
    if node.val == target:
      return node
    for child in node.children:
      result = self.search(child , target)
      if result:
        return result
    return None

  def display(self,node=None,level=0):
    if node is None:
      node = self.root

    print(" " * (level * 2) + str(node.val))

    for child in node.children:
      self.display(child, level + 1)

  def delete_node(self,target_val):
    if self.root.val == target_val:
      print("Cannot delete the root node")
      return
    parent_node = self.find_parent(self.root, target_val)
    if parent_node:
      parent_node.remove_child(target_val)
      print(f"Deleted Node {target_val}")
    else:
      print(f"Node {target_val} not found")

  def find_parent(self,node,target):
    for child in node.children:
      if child.val == target:
        return node
      result = self.find_parent(child, target)
      if result:
        return result
      return None

  def find_depth(self,node=None):
    if not node:
      node = self.root
    if not node.children:
      return 1
    return 1 + max(self.find_depth(child) for child in node.children)

  def main(self):
    while True:
      print("""
            1. Add element
            2. Delete element
            3. Display tree
            4. Search element
            5. Find depth
            6. Exit
            """)
      choice = int(input("Enter your choice: "))
      if choice == 1:
        parent_val = input("Enter the parent node value: ")
        child_val = input("Enter the child node value: ")
        self.add_child(parent_val, child_val)

      elif choice == 2:
        target = input("Enter the node value you want to search: ")
        self.delete_node(target)

      elif choice == 3:
        print("The Tree Structure")
        self.display()

      elif choice == 4:
        target = input("Enter the node value to search: ")
        result = self.search(self.root,target)
        if result:
          print(f"Node {target} found")
        else:
          print(f"Node {target} not found")

      elif choice == 5:
        depth = self.find_depth()
        print(f"The depth of the tree is {depth }")

      elif choice == 6:
        print("Exiting")
        break

      else:
        print("INVALID INPUT!!!")

tree = Tree("A")
tree.main()
````

### 1. **Binary Tree**

A **Binary Tree** is a hierarchical data structure where each node has at most two children: the **left child** and the **right child**. The basic operations include **insertion**, **deletion**, **searching**, and traversals like **DFS (Depth First Search)** and **BFS (Breadth First Search)**.

#### Characteristics:
1. **Root**: The top node of the tree.
2. **Node**: Contains a value and links to left and right children.
3. **Leaf**: A node with no children.
4. **Depth**: The number of edges from the root to a node.
5. **Height**: The longest path from the node to a leaf.

### 2. **Binary Search Tree (BST)**

A **Binary Search Tree (BST)** is a special kind of binary tree where:
- The **left child** contains values **less than** the parent node.
- The **right child** contains values **greater than** the parent node.
- This property allows efficient searching, insertion, and deletion operations.

#### Operations in BST:
1. **Insertion**: Adds a node while maintaining BST properties.
2. **Deletion**: Removes a node and restructures the tree to preserve properties.
3. **Search**: Finds a node based on the value.
4. **DFS Traversal**: In-order, Pre-order, and Post-order traversals.
5. **BFS Traversal**: Level-order traversal.

## Explanation of Traversals

1. **In-order DFS (Left, Root, Right)**: Visits nodes in ascending order.
2. **Pre-order DFS (Root, Left, Right)**: Useful for creating a copy of the tree.
3. **Post-order DFS (Left, Right, Root)**: Useful for deleting or freeing nodes.
4. **BFS (Level-order)**: Visits nodes level by level.

This implementation covers all key operations and traversal techniques in a Binary Search Tree.

````python
class Node:
  def __init__(self,val):
    self.val = val
    self.left = None
    self.right = None

  def __repr__(self):
    return str(self.val)

class Tree:
  def __init__(self):
    self.root = None

  def insert(self, val):
    new_node = Node(val)
    if not self.root:
      self.root = new_node

    else:
      queue = [self.root]
      while queue:
        current = queue.pop(0)
        if not current.left:
          current.left = new_node
          return
        else:
          queue.append(current.left)

        if not current.right:
          current.right = new_node
          return
        else:
          queue.append(current.right)

  def display(self,node= None, level=0,prefix='Root: '):
    if node is None:
      node = self.root
    if node:
      print(" " * (level * 4) + prefix + str(node.val))
      if node.left:
        self.display(node.left, level + 1 , 'L---')
      if node.right:
        self.display(node.right, level + 1, 'R---')

  def search(self,target):
    if not self.root:
      return None
    queue = [self.root]
    while queue:
      current = queue.pop(0)
      if current.val == target:
        return current
      if current.left:
        queue.append(current.left)
      if current.right:
        queue.append(current.right)
    return None

  def in_order(self,node=None):
    if node is None:
      return
    self.in_order(node.left)
    print(node.val,end=" ")
    self.in_order(node.right)

  def pre_order(self,node=None):
    if node is None:
      return
    print(node.val,end=" ")
    self.in_order(node.left)
    self.in_order(node.right)

  def post_order(self,node=None):
    if node is None:
      return
    self.in_order(node.left)
    self.in_order(node.right)
    print(node.val,end=" ")

  def BFS(self):
    if not self.root:
      return
    queue = [self.root]
    while queue:
      current = queue.pop(0)
      print(current.val, end =" ")
      if current.left:
        queue.append(current.left)
      if current.right:
        queue.append(current.right)

  def main(self):
    while True:
        print("""
        1. Insert element
        2. Search element
        3. In-order DFS Traversal
        4. Pre-order DFS Traversal
        5. Post-order DFS Traversal
        6. BFS Traversal
        7. Display Tree
        8. Exit
        """)

        choice = int(input("Enter your choice: "))
        if choice == 1:
          value = input("Enter the value to insert in the tree: ")
          self.insert(value)
          print(f"Value {value} inserted")

        elif choice == 2:
          search = input("Enter the value to search: ")
          self.search(search)

        elif choice == 3:
          print("IN-ORDER TRAVERSAL")
          self.in_order(self.root)
          print()

        elif choice == 4:
          print("PRE-ORDER TRAVERSAL")
          self.pre_order(self.root)
          print()

        elif choice == 5:
          print("POST-ORDER TRAVERSAL")
          self.post_order(self.root)
          print()

        elif choice == 6:
          print("BFS TRAVERSAL")
          self.BFS()
          print()

        elif choice == 7:
          print("Tree Structure: ")
          self.display()


        elif choice == 8:
          print("Exiting")
          break

        else:
          print("INVALID INPUT!!!")

tree = Tree()
tree.main()
````

````python
class Node:
  def __init__(self,val):
    self.val = val
    self.left = None
    self.right = None

  def __repr__(self):
    return str(self.val)

class Tree:
  def __init__(self):
    self.root = None

  def insert(self, val):
    new_node = Node(val)
    if not self.root:
      self.root = new_node
    else:
      self.insert_recursively(self.root,new_node)

  def insert_recursively(self,current,new_node):
    if new_node.val < current.val:
      if current.left is None:
        current.left = new_node
      else:
        self.insert_recursively(current.left,new_node)
    else:
      if current.right is None:
        current.right = new_node
      else:
        self.insert_recursively(current.right,new_node)

  def display(self,node= None, level=0,prefix='Root: '):
    if node is None:
      node = self.root
    if node:
      print(" " * (level * 4) + prefix + str(node.val))
      if node.left:
        self.display(node.left, level + 1 , 'L---')
      if node.right:
        self.display(node.right, level + 1, 'R---')

  def search(self,target):
    return self.search_recursively(self.root,target)

  def search_recursively(self,current,target):
      if current is None:
        return
      if current.val == target:
        return current
      if target < current.val:
        return self.search_recursively(current.left,target)
      else:
        return self.search_recursively(current.right,target)

  def delete(self, val):
      self.root = self.delete_recursively(self.root, val)

  def delete_recursively(self, current, target):
      if current is None:
          return None
      if target < current.val:
          current.left = self.delete_recursively(current.left, target)
      elif target > current.val:
          current.right = self.delete_recursively(current.right, target)
      else:
          if current.left is None and current.right is None:
              return None
          if current.left is None:
              return current.right
          if current.right is None:
              return current.left
          successor = self.find_min(current.right)
          current.val = successor.val
          current.right = self.delete_recursively(current.right, successor.val)
      return current

  def find_min(self, node):
      current = node
      while current.left is not None:
          current = current.left
      return current



  def in_order(self,node=None):
    if node is None:
      return
    self.in_order(node.left)
    print(node.val,end=" ")
    self.in_order(node.right)

  def pre_order(self,node=None):
    if node is None:
      return
    print(node.val,end=" ")
    self.in_order(node.left)
    self.in_order(node.right)

  def post_order(self,node=None):
    if node is None:
      return
    self.in_order(node.left)
    self.in_order(node.right)
    print(node.val,end=" ")

  def BFS(self):
    if not self.root:
      return
    queue = [self.root]
    while queue:
      current = queue.pop(0)
      print(current.val, end =" ")
      if current.left:
        queue.append(current.left)
      if current.right:
        queue.append(current.right)

  def main(self):
    while True:
        print("""
        1. Insert element
        2. Search element
        3. In-order DFS Traversal
        4. Pre-order DFS Traversal
        5. Post-order DFS Traversal
        6. BFS Traversal
        7. Display Tree
        8. Delete Element
        9. Exit
        """)

        choice = int(input("Enter your choice: "))
        if choice == 1:
          value = input("Enter the value to insert in the tree: ")
          self.insert(value)
          print(f"Value {value} inserted")

        elif choice == 2:
          search = input("Enter the value to search: ")
          result = self.search(search)
          if result:
            print(f"Value {search} found")
          else:
            print(f"Value {search} not found")


        elif choice == 3:
          print("IN-ORDER TRAVERSAL")
          self.in_order(self.root)
          print()

        elif choice == 4:
          print("PRE-ORDER TRAVERSAL")
          self.pre_order(self.root)
          print()

        elif choice == 5:
          print("POST-ORDER TRAVERSAL")
          self.post_order(self.root)
          print()

        elif choice == 6:
          print("BFS TRAVERSAL")
          self.BFS()
          print()

        elif choice == 7:
          print("Tree Structure: ")
          self.display()

        elif choice == 8:
          delete = input('Enter the value you want to delete from the tree: ')
          result = self.delete(delete)
          print(f"{delete} deleted")

        elif choice == 9:
          print("Exiting")
          break

        else:
          print("INVALID INPUT!!!")

tree = Tree()
tree.main()
````

###  **Explanation of Functionalities**

1. **Insert**:
   - Adds a node to the BST while maintaining the BST property.
   - Uses a recursive function to find the correct position based on the value.

2. **Search**:
   - Searches for a node based on the value.
   - Efficient in BST due to ordered structure.

3. **Delete**:
   - Deletes a node from the BST.
   - Three cases for deletion:
     1. Node with no children.
     2. Node with one child.
     3. Node with two children (finds the in-order successor).

4. **In-order Traversal** (DFS):
   - **Left  Root  Right**
   - Outputs the values in **sorted order** for a BST.

5. **Pre-order Traversal** (DFS):
   - **Root  Left  Right**

6. **Post-order Traversal** (DFS):
   - **Left  Right  Root**

7. **BFS** (Level-order Traversal):
   - Visits nodes level by level using a queue.

8. **Display**:
   - Displays the tree structure with proper indentation to visualize the hierarchy.
