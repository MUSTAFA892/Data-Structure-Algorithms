A **Heap** is a special type of **complete binary tree** that satisfies the **heap property**. It is primarily used to implement **priority queues**, where we want quick access to the highest or lowest priority element.

---

## ✅ What is a Complete Binary Tree?

A **complete binary tree** is a tree in which all levels are completely filled except possibly the last, and the last level has all nodes as far left as possible.

Visual Example:

```
    10
   /  \
  20   30
 / \
40 50
```

This structure ensures **balanced shape**, which helps in efficient storage and access when implemented using an **array**.

---

## 💡 Heap Property

Every heap must follow one of these two properties:

1. **Max-Heap Property**:

   * Every parent node's value is **greater than or equal to** its children.
   * Root always contains the **maximum** value.

2. **Min-Heap Property**:

   * Every parent node's value is **less than or equal to** its children.
   * Root always contains the **minimum** value.

---

## 🧱 Types of Heap

### 1. **Max Heap**

* Property: `Parent ≥ Children`
* Example:

```
         90
       /    \
     15      10
    /  \     /
   7   12   2
```

* Array representation: `[90, 15, 10, 7, 12, 2]`
* Application: **Priority queues**, **heap sort**

---

### 2. **Min Heap**

* Property: `Parent ≤ Children`
* Example:

```
         2
       /   \
     7      10
    / \     /
  15  12   90
```

* Array representation: `[2, 7, 10, 15, 12, 90]`
* Application: **Shortest path algorithms (Dijkstra's)**, **min-priority queues**

---

## ⚙️ Basic Heap Operations

### 1. **Insertion**

* Insert at the **end of the array** (i.e., as a leaf in the tree).
* Then **"heapify up"** to restore the heap property.

#### Example (Max Heap):

Inserting `100` into `[90, 15, 10, 7, 12, 2]`:

1. Insert 100 at the end: `[90, 15, 10, 7, 12, 2, 100]`
2. Compare 100 with its parent (10), swap.
3. Compare with next parent (90), swap.
4. Final: `[100, 15, 90, 7, 12, 2, 10]`

---

### 2. **Deletion (usually root)**

* Remove the **root element**.
* Replace it with the **last element** in the array.
* **"Heapify down"** to maintain heap property.

#### Example (Max Heap):

Deleting `100` from `[100, 15, 90, 7, 12, 2, 10]`:

1. Replace 100 with last element → `10`: `[10, 15, 90, 7, 12, 2]`
2. Compare with children: 90 > 10 → swap
3. Final: `[90, 15, 10, 7, 12, 2]`

---

### 3. **Peek (Get Max/Min)**

* Just return the root element.
* Time Complexity: **O(1)**

---

### 4. **Heapify (Build a Heap)**

* Converts an arbitrary array into a heap in **O(n)** time using bottom-up approach.

---

## 🧮 Array Representation of a Heap

In an array-based binary heap:

* Parent index: `i`
* Left child: `2i + 1`
* Right child: `2i + 2`
* Parent of `i`: `(i - 1) // 2`

Example (Max Heap):

Array: `[90, 15, 10, 7, 12, 2]`

```
         90 (i=0)
       /      \
    15 (1)    10 (2)
   /    \     /
 7(3) 12(4) 2(5)
```

## 🧑‍⚕️ Real-World Example: Hospital Emergency Room

* Each patient has a priority (higher = more critical).
* We use a **Max Heap** to always serve the most critical patient.

### Actions:

* New patient → Insert and heapify up.
* Serve patient → Remove max and heapify down.
* Check who's next → Peek root.

## ⏱️ Time and Space Complexity

| Operation      | Time Complexity |
| -------------- | --------------- |
| Insert         | O(log n)        |
| Delete (Root)  | O(log n)        |
| Peek (Max/Min) | O(1)            |
| Build Heap     | O(n)            |
| Space (Array)  | O(n)            |

## 🧠 Applications of Heap

1. **Priority Queues** – Used in OS scheduling, load balancing.
2. **Heap Sort** – Sorting in O(n log n) time.
3. **Dijkstra’s Algorithm** – For shortest paths using Min Heap.
4. **Median Maintenance** – Two heaps (Max + Min) to track median efficiently.
5. \**AI Pathfinding (A*)\*\*, **Job Scheduling**, **Data Stream Analysis**




### 🔑 **Heap Terminologies (General)**

1. **Heap**

   * A **complete binary tree** where all levels are filled except possibly the last, and the last level is filled from left to right.
   * Implemented efficiently using arrays.

2. **Min Heap**

   * A binary heap where the **parent node is always smaller** than its children.
   * The **smallest** element is always at the **root**.
   * Used in **priority queues**, **Dijkstra’s algorithm**, **event-driven simulators**, etc.

3. **Max Heap**

   * A binary heap where the **parent node is always larger** than its children.
   * The **largest** element is at the **root**.
   * Commonly used in **heap sort**, **priority queues**, **real-time systems**, etc.

4. **Root Node**

   * The first element in the heap.
   * For Min Heap: smallest element.
   * For Max Heap: largest element.

5. **Heapify**

   * A process to maintain the heap property after insertion or deletion.
   * Two types:

     * **Heapify-up (Bubble-up)**: Used after insertion.
     * **Heapify-down (Bubble-down)**: Used after deletion.

6. **Complete Binary Tree**

   * A tree where every level except possibly the last is completely filled, and the last level is filled left to right.

7. **Priority Queue**

   * A data structure where each element has a priority. Heaps are a typical way to implement them.



### 🐍 `heapq` Module in Python

Python’s `heapq` supports **Min Heap** by default.

#### Operations:

| Operation                      | Description                                     |
| ------------------------------ | ----------------------------------------------- |
| `heapq.heapify(lst)`           | Converts a list into a heap (in-place).         |
| `heapq.heappush(heap, val)`    | Pushes a value while maintaining heap property. |
| `heapq.heappop(heap)`          | Pops the smallest value (root) from heap.       |
| `heapq.heappushpop(heap, val)` | Push then pop smallest in one call.             |
| `heapq.nlargest(n, heap)`      | Returns n largest values from heap.             |
| `heapq.nsmallest(n, heap)`     | Returns n smallest values from heap.            |

---

### ⚙️ How to Create Max Heap using `heapq`?

`heapq` only supports **Min Heap**, so to simulate a **Max Heap**:

* Insert the **negative of values**.
* When accessing values, **negate them again**.

Example:

```python
import heapq
heap = []
heapq.heappush(heap, -10)  # simulate max heap
heapq.heappush(heap, -1)
heapq.heappush(heap, -20)
max_val = -heapq.heappop(heap)  # convert back to positive
```


### Min Heap

````python
import heapq

class MinHeapArray:
    def __init__(self):
        self.heap = []

    def insert(self):
        n = int(input("How many elements you want to insert into the Min Heap? "))
        print()
        for i in range(1, n+1):
            val = int(input(f"Enter element {i}: "))
            heapq.heappush(self.heap, val)
        print("All elements inserted into the Min Heap successfully.\n")

    def delete(self):
        if self.heap:
            removed = heapq.heappop(self.heap)
            print(f"Removed element: {removed}\n")
        else:
            print("Heap is empty.\n")

    def traverse(self):
        print(f"Current Min Heap: {self.heap}\n")

    def search(self):
        val = int(input("Enter the value to search in Min Heap: "))
        if val in self.heap:
            print(f"Element {val} found at index {self.heap.index(val)}\n")
        else:
            print("Element not found in the heap.\n")

    def menu(self):
        while True:
            print("""
----- Min Heap Operations -----
1. Insert
2. Delete Min
3. Traverse
4. Search
5. Exit
""")
            choice = int(input("Enter your choice: "))
            print()

            if choice == 1:
                self.insert()
            elif choice == 2:
                self.delete()
            elif choice == 3:
                self.traverse()
            elif choice == 4:
                self.search()
            elif choice == 5:
                break
            else:
                print("Invalid choice\n")


min_heap = MinHeapArray()
min_heap.menu()
````

## Max Heap

````python
import heapq

class MaxHeapArray:
    def __init__(self):
        self.heap = []

    def insert(self):
        n = int(input("How many elements you want to insert into the Max Heap? "))
        print()
        for i in range(1, n+1):
            val = int(input(f"Enter element {i}: "))
            heapq.heappush(self.heap, -val)
        print("All elements inserted into the Max Heap successfully.\n")

    def delete(self):
        if self.heap:
            removed = -heapq.heappop(self.heap)
            print(f"Removed element: {removed}\n")
        else:
            print("Heap is empty.\n")

    def traverse(self):
        real_heap = [-x for x in self.heap]
        print(f"Current Max Heap: {real_heap}\n")

    def search(self):
        val = int(input("Enter the value to search in Max Heap: "))
        if -val in self.heap:
            idx = self.heap.index(-val)
            print(f"Element {val} found at index {idx} (as -{val} in list)\n")
        else:
            print("Element not found in the heap.\n")

    def menu(self):
        while True:
            print("""
----- Max Heap Operations -----
1. Insert
2. Delete Max
3. Traverse
4. Search
5. Exit
""")
            choice = int(input("Enter your choice: "))
            print()

            if choice == 1:
                self.insert()
            elif choice == 2:
                self.delete()
            elif choice == 3:
                self.traverse()
            elif choice == 4:
                self.search()
            elif choice == 5:
                break
            else:
                print("Invalid choice\n")


max_heap = MaxHeapArray()
max_heap.menu()
````
