### String Data Structure

A **string** is a sequence of characters used to represent text in programming. It is one of the fundamental data structures in most programming languages and is often implemented as an array of characters. Strings can hold letters, numbers, symbols, and spaces, making them versatile for various applications, such as text processing, data representation, and user interface elements.

Strings are usually immutable, meaning that once created, their contents cannot be changed. However, many programming languages provide methods to manipulate strings, allowing for the creation of new strings based on existing ones.

### Common Operations on Strings

1. **Concatenation**:
   - **Description**: Joining two or more strings together to form a new string.
   - **Example**: `"Hello" + " " + "World"` results in `"Hello World"`.

2. **Substring Extraction**:
   - **Description**: Retrieving a portion of a string based on specified indices.
   - **Example**: From the string `"Hello World"`, extracting the substring from index 6 to 11 gives `"World"`.

3. **Searching**:
   - **Description**: Finding the position of a substring within a string.
   - **Example**: Checking if the substring `"Hello"` exists in `"Hello World"` returns `True` and its starting index, which is `0`.

4. **Length**:
   - **Description**: Determining the number of characters in a string.
   - **Example**: The length of the string `"Hello"` is `5`.

5. **Replacement**:
   - **Description**: Replacing occurrences of a substring with another substring.
   - **Example**: Replacing `"World"` with `"Everyone"` in `"Hello World"` results in `"Hello Everyone"`.

6. **Splitting**:
   - **Description**: Dividing a string into a list (or array) of substrings based on a delimiter.
   - **Example**: Splitting `"apple,banana,cherry"` by `","` results in the list `["apple", "banana", "cherry"]`.



```python
class StringOperations:
    def __init__(self, text):
        self.text = text

    def concatenate(self, new_text):
        self.text += new_text
        print(f"Concatenated String: {self.text}")

    def substring(self, start, end):
        sub_text = self.text[start:end]
        print(f"Substring: {sub_text}")

    def search(self, substring):
        position = self.text.find(substring)
        if position != -1:
            print(f"Substring '{substring}' found at index: {position}")
        else:
            print(f"Substring '{substring}' not found.")

    def length(self):
        print(f"Length of the string: {len(self.text)}")

# Example usage
initial_text = input("Enter initial string: ")
string_ops = StringOperations(initial_text)

while True:
    print("""
    1. Concatenate
    2. Substring
    3. Search
    4. Length


    5. Exit
    """)
    choice = int(input("Enter your choice: "))

    if choice == 1:
        new_text = input("Enter string to concatenate: ")
        string_ops.concatenate(new_text)

    elif choice == 2:
        start = int(input("Enter start index: "))
        end = int(input("Enter end index: "))
        string_ops.substring(start, end)

    elif choice == 3:
        substring = input("Enter substring to search for: ")
        string_ops.search(substring)

    elif choice == 4:
        string_ops.length()

    elif choice == 5:
        print("Exiting...")
        break

    else:
        print("Invalid choice. Please try again.")
```
